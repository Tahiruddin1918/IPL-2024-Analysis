```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
def fix_date_format(date_str):
    date_str = date_str.replace("IST", "").strip() + ", 2024"
    
    return pd.to_datetime(date_str, format="%b, %a %d , %I:%M %p, %Y")

df["Date & Time"] = df["Date & Time"].apply(fix_date_format)

df.head()
```

```python
df['Team 1 Score'] = df['Team 1 Score'].str.split('/').str[0].astype(int)
df['Team 2 Score'] = df['Team 2 Score'].str.split('/').str[0].astype(int)

df[['Team 1 Score','Team 2 Score']].head()
```

```python
df['Winning Margin Value'] = df['winning_margin'].str.extract('(\d+)').astype(int)
df['Winning Margin Type'] = df['winning_margin'].str.extract("(RUNS|WICKETS)")
df[['winning_margin','Winning Margin Value','Winning Margin Type']].head()
```

```python
df.to_csv('ipl_cleaned_data.csv', index=False)
```

```python
df.head()
```

```python
team_replacements = {
    'Kolkata Knight Riders': 'KKR',
    'Sunrisers Hyderabad': 'SRH',
    'Chennai Super Kings': 'CSK',
    'Delhi Capitals': 'DC',
    'Mumbai Indians': 'MI',
    'Royal Challengers Bengaluru': 'RCB',
    'Lucknow Super Giants': 'LSG',
    'Punjab Kings': 'PK',
    'Gujarat Titans': 'GT',
    'Rajasthan Royals': 'RR'
}
Winners_team = {key.upper(): value for key, value in team_replacements.items()}

df[['Team 1', 'Team 2']] = df[['Team 1', 'Team 2']].replace(team_replacements)
df['winner'] = df['winner'].replace(Winners_team)
```
